# Fitting a Gaussian w/ Least Squares

```julia
t = 0.:0.02:1.0
```

```julia
gaussian(x; x0=0., σ=1.) = exp(-((x - x0)/σ)^2)
```

```julia
ytrue = gaussian.(t, x0=0.5, σ=0.2);
p = plot(t, ytrue, linewidth=2.5, label="truth")

d = Normal(0.0, 0.2)
yobs = ytrue .+ rand(d, length(ytrue));

plot!(p, t, yobs, markershape=:circle, linewidth=0, label="samples")
```

```julia
ytrue = gaussian.(t, x0=0.5, σ=0.1);
yint = sum(ytrue)
p = plot(ytrue, t, linewidth=2.5, label="t=0")
ytrue = gaussian.(t, x0=0.5, σ=0.2);
ytrue *= yint/sum(ytrue)
plot!(p, ytrue, t, linewidth=2.5, label="t=1", legendfontsize=12, xlabel="C", ylabel="z", size=(250,400))
```
